```python
# 트렌드 확인 
from pytrends.request import TrendReq
import pandas as pd
import datetime
# Only need to run this once, the rest of requests will use the same session.
def strftime(date) :
    return date.strftime("%Y-%m-%d")

pytrend = TrendReq()

for _vcoin in volume_coin :
    
    keyward = marketlist_all[_vcoin] + ' coin'
    print(keyward)
    kw_list = [keyward]
    pytrend.build_payload(kw_list, cat=0, timeframe='now 7-d', geo='', gprop='')
    data = pytrend.interest_over_time()
    data.reset_index(level=0, inplace=True)
    data['date'] = data['date'].apply(strftime)
    temp = { 'date' : [], 'count' : []}
    for i in range(1, 8) :
        _time = datetime.datetime.today() - datetime.timedelta(i)
        _time = _time.strftime("%Y-%m-%d")
        temp['date'].append(_time)
        temp['count'].append(data[data['date'] == _time][keyward].sum())
    temp = pd.DataFrame.from_dict(temp)
    print(temp[::-1])
```
